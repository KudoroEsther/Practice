```python
df1 = pd.read_csv('/content/drive/MyDrive/keamanan data/emotion_dataset/Emotion_classify_Data.csv')
df1
```

```python
df1.drop('Emotion', axis=1, inplace=True)
```

```python
def determine_emotion(comment):
    emotions_dict = {
        'empty': ['empty', 'void', 'hollow'],
        'sadness': ['sad', 'melancholy', 'depressed'],
        'enthusiasm': ['enthusiastic', 'excited', 'eager'],
        'neutral': ['neutral', 'indifferent', 'unbiased'],
        'worry': ['worry', 'anxiety', 'concern'],
        'surprise': ['surprise', 'astonishment', 'shock'],
        'love': ['love', 'affection', 'adoration'],
        'fun': ['fun', 'joyful', 'amusing'],
        'hate': ['hate', 'detest', 'loathe'],
        'happiness': ['happy', 'joy', 'content'],
        'boredom': ['boredom', 'tedium', 'monotony'],
        'relief': ['relief', 'ease', 'comfort'],
        'anger': ['angry', 'rage', 'outrage']
    }

    for emotion, keywords in emotions_dict.items():
        for keyword in keywords:
            if keyword in comment:
                return emotion
    return "neutral"  # default to neutral if no emotion is detected

for comment in df1['Comment']:
    df1['Emotion'] = df1['Comment'].apply(lambda comment: determine_emotion(comment.lower()))
df1
```

```python
df2.drop(['Unnamed: 0', 'label'], axis=1, inplace=True)
```

```python
def determine_emotion(comment):
    emotions_dict = {
        'empty': ['empty', 'void', 'hollow'],
        'sadness': ['sad', 'melancholy', 'depressed'],
        'enthusiasm': ['enthusiastic', 'excited', 'eager'],
        'neutral': ['neutral', 'indifferent', 'unbiased'],
        'worry': ['worry', 'anxiety', 'concern'],
        'surprise': ['surprise', 'astonishment', 'shock'],
        'love': ['love', 'affection', 'adoration'],
        'fun': ['fun', 'joyful', 'amusing'],
        'hate': ['hate', 'detest', 'loathe'],
        'happiness': ['happy', 'joy', 'content'],
        'boredom': ['boredom', 'tedium', 'monotony'],
        'relief': ['relief', 'ease', 'comfort'],
        'anger': ['angry', 'rage', 'outrage']
    }

    for emotion, keywords in emotions_dict.items():
        if any(keyword in comment.lower() for keyword in keywords):  # Use `any` for efficient detection
            return emotion

    return "neutral"  # Default to neutral if no emotion is detected

df2['Emotion'] = df2['text'].apply(lambda comment: determine_emotion(comment.lower()))
df2
```
