## Shopping List Master

Create an empty list.

Ask the user to enter 3 shopping items (one by one).

Add them to the list.

Display the list as a single string, separated by commas.

```python
# Shopping list
shop = []
for i in range(3):
    item = input('Enter an item: ')
    shop.append(item)
print("Shopping List")

print(', '.join(shop)) # converts the list to string
```

## Name Organizer
Ask the user to enter 5 names (separated by spaces).

Convert all names to lowercase.

Sort the list alphabetically.

Display them one name per line.

```python
names = input("Enter 5 names separated by spaces: ").lower().split()
names.sort()
for i in names:
    print(i)
```

## Student Score Tracker

## Word Analyzer
Ask the user to input a word.

Print the length of the word.

Check if it is all uppercase, all lowercase, or title case.

Reverse the word using slicing.

```python
word = input("Enter a word: ")
print(len(word))

if word.isupper():
    print(f"'{word}' is upper case")
elif word.islower():
    print(f"'{word}' is lowercase")
elif word.istitle():
    print(f"'{word}' is title case")

print(f"Reveresed word {word[::-1]}")
```

## List Manipulation
Create a list of five cities.

Replace the third city with a new one (entered by the user).

Remove the last city.

Add a new city to the beginning of the list.

Print the updated list

```python
cities = ['Lagos', 'Yonsei', 'Okinawa', 'LA', 'Qatar']

replac = input("Enter a city: ")
cities[2] = replac

print(cities)

cities.pop()
print(cities)

cities.insert(0, 'Manila')
print(cities)
```
