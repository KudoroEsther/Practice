## Practice Questions
Create a 3x3 matrix of random integers and:

   - Multiply it by 2

   - Transpose it

   - Find the determinant

   - Check if it’s invertible

   - If yes, find the inverse


```python
ran = np.random.randint(1,10, size=(3,3))
print(ran)
ran.T
print(f"\nTranspose \n{ran.T}")

print(f"\nDeterminant: \n{np.linalg.det(ran)}")
print(f"\nNon invertible: {(np.linalg.det(ran))!=0}")
print(f"\nInverse: \n{np.linalg.inv(ran)}")
```
